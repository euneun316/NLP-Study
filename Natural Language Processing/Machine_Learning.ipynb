## 02) 머신 러닝 훑어보기

### 1.머신 러닝 모델의 *평가*

데이터를 훈련용, 검증용, 테스트용 이렇게 세 가지로 분리
- 훈련 데이터 : 머신 러닝 모델을 학습하는 용도 : `문제지`
- 테스트 데이터 : 학습한 머신 러닝 모델의 성능을 평가하기 위한 용도 : `수능시험`
- 검증용 데이터 : `모의고사`
  - 모델의 성능을 평가하기 위한 용도가 아니라 모델의 성능을 조정하기 위한 용도
  - 훈련 데이터에 과적합(overfitting) 이 되고 있는지 판단하거나 하이퍼파라미터의 조정을 위한 용도
---
- 하이퍼 파라미터(초매개변수) : 모델의 성능에 영향을 주는 사람이 값을 지정하는 변수. == **사용자가 직접 정해줄 수 있는 변수** <br>
ex) 선형 회귀의 경사 하강법에서의 학습률, 딥러닝에서 뉴런의 수나 층의 수
- 매개변수 : 가중치와 편향. 학습을 하는 동안 값이 계속해서 변하는 수 == **모델이 학습하는 과정에서 얻어지는 값**
---
1. 훈련 데이터로 훈련 시킨 모델은 **검증용 데이터**를 사용해 정확도 검증하며 **하이퍼 파라미터**를 **튜닝(tuning)**한다. <br>
2. 튜닝하는 과정에서 모델은 검증용 데이터의 정확도를 높이는 방향으로 점차적으로 수정
3. 튜닝 과정을 모두 끝내고 모델의 최종 평가는 **테스트 데이터**로 수행


### 2.분류(Classification)와 회귀(Regression)

회귀 : 선형 회귀 <br>
분류 : 로지스틱 회귀 

#### 1) 이진 분류 문제(Binary Classification)
- 주어진 입력에 대해서 두 개의 선택지 중 하나의 답을 선택
- ex) 시험 성적보고, 합격, 불합격인지 판단하는 문제
- ex) 메일을 보고, 스팸인지 정상 메일인지 판단하는 문제

#### 2) 다중 클래스 분류(Multi-class Classification)
- 세 개 이상의 선택지 중에서 답을 선택해야 하는 경우
- ex) 과학, 영어, IT, 학습지, 만화라는 5가지 레이블 중 적절한 분야를 선택하는 문제

#### 3) 회귀 문제(Regression)
- 정답이 몇 개의 정해진 선택지 중에서 정해져 있는 경우가 아니라, 어떠한 **연속적인 값의 범위** 내에서 예측값이 나오는 경우
- 특정 값의 범위 내에서는 어떤 숫자도 나올 수 있다.
- ex) 시계열 데이터(Time Series Data)를 이용한 주가 예측, 생산량 예측, 지수 예측
- ex) 부동산 가격이 7억 8,456만 3,450원, 또는 8억 1257만 300원으로 예측하는 경우

### 3.지도 학습과 비지도 학습

#### 1) 지도 학습(Supervised Learning)

- 레이블(Label)이라는 정답과 함께 학습
- `레이블` == `실제값` == `y`

#### 2) 비지도 학습(Unsupervised Learning)

- 데이터에 별도의 레이블이 없이 학습하는 것
- ex) 토픽 모델링 알고리즘 LSA, LDA - 텍스트 마이닝 기법

#### 3) 자기지도 학습(Self-Supervised Learning, SSL)

- 레이블이 없는 데이터가 주어지면, 모델이 학습을 위해서 스스로 데이터로부터 레이블을 만들어서 학습
- Word2Vec과 같은 워드 임베딩, BERT와 같은 언어 모델

- 인공 신경망은 독립 변수, 종속 변수, 가중치, 편향 등을 행렬 연산을 통해 연산
- 훈련 데이터를 행렬로 표현
- 독립 변수 의 행렬을 X라 할 때 / 독립 변수의 개수가 n개 / 데이터의 개수가 m개

### 5.혼동 행렬(Confusion Matrix)

**정확도(Accuracy)** : 맞춘 문제수를 전체 문제수로 나눈 값 <br>
=> 맞춘 결과와 틀린 결과에 대한 세부적인 내용을 알려주지는 않음 <br>
이를 위해서 사용하는 것이 **혼동 행렬(Confusion Matrix)** <br><br>

image.png

- True Positive(TP) : 실제 True인 정답을 True라고 예측 (정답)
- True Negative(TN) : 실제 False인 정답을 False라고 예측 (정답)
- False Positive(FP) : 실제 False인 정답을 True라고 예측 (오답)
- False Negative(FN) : 실제 True인 정답을 False라고 예측 (오답)

  (앞은 예측의 맞음 틀림/ 뒤는 모델의 예측)

---

- 1종 오류(FP) : 귀무가설이 참인데도 불구하고 귀무가설을 기각하고 대립가설을 채택하는 오류
- 2종 오류(FN) : 귀무가설이 거짓인데도 불구하고 귀무가설을 채택하는 오류
  (보통 1종 오류가 2종 오류보다 위험)

- [통계적 가설 검정의 오류]

image.png


1) 정밀도(Precision)
- 모델이 True라고 분류한 것 중에서 실제 True인 것의 비율<br><br>
$$
TP \over TP + FP
$$

2) 재현율(Recall)
- 실제 True인 것 중에서 모델이 True라고 예측한 것의 비율<br><br>
$$
TP \over TP + FN
$$

Precision이나 Recall은 모두 실제 True인 정답을 모델이 True라고 예측한 경우 **TP**에 관심이 있다.

3) 정확도(Accuracy)

- 전체 예측한 데이터 중에서 정답을 맞춘 것에 대한 비율 <br><br>
$$
TP+TN \over TP + FN + FP + TN
$$
<br>

- Accuracy로 성능을 예측하는 것이 적절하지 않은 때도 있다.
- ex) 메일 100개 중 스팸 메일은 5개,  모두 정상 메일이라고 예측하면, 스팸 메일은 하나도 못찾음
- 이런 경우 F1-Score를 사용

### 6.과적합(Overfitting)과 과소 적합(Underfitting)

 과적합(Overfitting) : 훈련 데이터를 과하게 학습한 경우
 - 훈련 데이터에 대해서만 과하게 학습하면 성능 측정을 위한 데이터인 테스트 데이터나 실제 서비스에서는 정확도가 좋지 않은 현상이 발생
 - 훈련 데이터에 대해서는 오차가 낮지만, 테스트 데이터에 대해서는 오차가 커진다.
 - 훈련 데이터에 대한 훈련 횟수에 따른 훈련 데이터의 오차와 테스트 데이터의 오차

 image.png

 - y축은 오차(loss)
 - X축의 에포크(epoch)는 전체 훈련 데이터에 대한 훈련 횟수
 - 테스트 데이터의 오차가 증가하기 전이나, 정확도가 감소하기 전에 훈련을 멈추는 것이 바람직 함

과소적합(Underfitting) : 테스트 데이터의 성능이 올라갈 여지가 있음에도 훈련을 덜 한 상태
- 과소 적합은 훈련 자체를 너무 적게한 상태이므로 훈련 데이터에 대해서도 정확도가 낮다
- 머신 러닝에서 학습 또는 훈련 == 적합(fitting)이라고도 부름
- 모델이 주어진 데이터에 대해서 적합해져가는 과정
- 딥 러닝을 할 때는 과적합을 막을 수 있는 드롭 아웃(Dropout), 조기 종료(Early Stopping)과 같은 몇 가지 방법이 존재
---

[과적합 방지를 고려한 일반적인 딥 러닝 모델의 학습 과정]

1. 주어진 데이터를 훈련 데이터, 검증 데이터, 테스트 데이터로 나눈다. 가령, 6:2:2 비율로 나눌 수 있다.
2. 훈련 데이터로 모델을 학습한다. (에포크 +1)
3. 검증 데이터로 모델을 평가하여 검증 데이터에 대한 정확도와 오차(loss)를 계산한다.
4. 검증 데이터의 오차가 증가하였다면 과적합 징후이므로 학습 종료 후 Step 5로 이동, 아니라면 Step 2.로 재이동한다.
5. 모델의 학습이 종료되었으니 테스트 데이터로 모델을 평가한다.
