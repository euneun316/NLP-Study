## 02) 머신 러닝 훑어보기

### 1.머신 러닝 모델의 *평가*

데이터를 훈련용, 검증용, 테스트용 이렇게 세 가지로 분리
- 훈련 데이터 : 머신 러닝 모델을 학습하는 용도 : `문제지`
- 테스트 데이터 : 학습한 머신 러닝 모델의 성능을 평가하기 위한 용도 : `수능시험`
- 검증용 데이터 : `모의고사`
  - 모델의 성능을 평가하기 위한 용도가 아니라 모델의 성능을 조정하기 위한 용도
  - 훈련 데이터에 과적합(overfitting) 이 되고 있는지 판단하거나 하이퍼파라미터의 조정을 위한 용도
---
- 하이퍼 파라미터(초매개변수) : 모델의 성능에 영향을 주는 사람이 값을 지정하는 변수. == **사용자가 직접 정해줄 수 있는 변수** <br>
ex) 선형 회귀의 경사 하강법에서의 학습률, 딥러닝에서 뉴런의 수나 층의 수
- 매개변수 : 가중치와 편향. 학습을 하는 동안 값이 계속해서 변하는 수 == **모델이 학습하는 과정에서 얻어지는 값**
---
1. 훈련 데이터로 훈련 시킨 모델은 **검증용 데이터**를 사용해 정확도 검증하며 **하이퍼 파라미터**를 **튜닝(tuning)**한다. <br>
2. 튜닝하는 과정에서 모델은 검증용 데이터의 정확도를 높이는 방향으로 점차적으로 수정
3. 튜닝 과정을 모두 끝내고 모델의 최종 평가는 **테스트 데이터**로 수행


### 2.분류(Classification)와 회귀(Regression)

회귀 : 선형 회귀 <br>
분류 : 로지스틱 회귀 

#### 1) 이진 분류 문제(Binary Classification)
- 주어진 입력에 대해서 두 개의 선택지 중 하나의 답을 선택
- ex) 시험 성적보고, 합격, 불합격인지 판단하는 문제
- ex) 메일을 보고, 스팸인지 정상 메일인지 판단하는 문제

#### 2) 다중 클래스 분류(Multi-class Classification)
- 세 개 이상의 선택지 중에서 답을 선택해야 하는 경우
- ex) 과학, 영어, IT, 학습지, 만화라는 5가지 레이블 중 적절한 분야를 선택하는 문제

#### 3) 회귀 문제(Regression)

- 특정 값의 범위 내에서는 어떤 숫자도 나올 수 있다.
- ex) 시계열 데이터(Time Series Data)를 이용한 주가 예측, 생산량 예측, 지수 예측
- ex) 부동산 가격이 7억 8,456만 3,450원, 또는 8억 1257만 300원으로 예측하는 경우

### 5.혼동 행렬(Confusion Matrix)

**정확도(Accuracy)** : 맞춘 문제수를 전체 문제수로 나눈 값 <br>
=> 맞춘 결과와 틀린 결과에 대한 세부적인 내용을 알려주지는 않음 <br>
이를 위해서 사용하는 것이 **혼동 행렬(Confusion Matrix)** <br><br>

image.png

- True Positive(TP) : 실제 True인 정답을 True라고 예측 (정답)
- True Negative(TN) : 실제 False인 정답을 False라고 예측 (정답)
- False Positive(FP) : 실제 False인 정답을 True라고 예측 (오답)
- False Negative(FN) : 실제 True인 정답을 False라고 예측 (오답)

  (앞은 예측의 맞음 틀림/ 뒤는 모델의 예측)

---

- 1종 오류(FP) : 귀무가설이 참인데도 불구하고 귀무가설을 기각하고 대립가설을 채택하는 오류
- 2종 오류(FN) : 귀무가설이 거짓인데도 불구하고 귀무가설을 채택하는 오류
  (보통 1종 오류가 2종 오류보다 위험)

- [통계적 가설 검정의 오류]

image.png


1) 정밀도(Precision)
- 모델이 True라고 분류한 것 중에서 실제 True인 것의 비율<br><br>
$$
TP \over TP + FP
$$

2) 재현율(Recall)
- 실제 True인 것 중에서 모델이 True라고 예측한 것의 비율<br><br>
$$
TP \over TP + FN
$$

Precision이나 Recall은 모두 실제 True인 정답을 모델이 True라고 예측한 경우 **TP**에 관심이 있다.

3) 정확도(Accuracy)

- 전체 예측한 데이터 중에서 정답을 맞춘 것에 대한 비율 <br><br>
$$
TP+TN \over TP + FN + FP + TN
$$
<br>

- Accuracy로 성능을 예측하는 것이 적절하지 않은 때도 있다.
- ex) 메일 100개 중 스팸 메일은 5개,  모두 정상 메일이라고 예측하면, 스팸 메일은 하나도 못찾음
- 이런 경우 F1-Score를 사용

### 6.과적합(Overfitting)과 과소 적합(Underfitting)

 과적합(Overfitting) : 훈련 데이터를 과하게 학습한 경우
 - 훈련 데이터에 대해서만 과하게 학습하면 성능 측정을 위한 데이터인 테스트 데이터나 실제 서비스에서는 정확도가 좋지 않은 현상이 발생
 - 훈련 데이터에 대해서는 오차가 낮지만, 테스트 데이터에 대해서는 오차가 커진다.
 - 훈련 데이터에 대한 훈련 횟수에 따른 훈련 데이터의 오차와 테스트 데이터의 오차

 image.png

 - y축은 오차(loss)
 - X축의 에포크(epoch)는 전체 훈련 데이터에 대한 훈련 횟수
 - 테스트 데이터의 오차가 증가하기 전이나, 정확도가 감소하기 전에 훈련을 멈추는 것이 바람직 함

과소적합(Underfitting) : 테스트 데이터의 성능이 올라갈 여지가 있음에도 훈련을 덜 한 상태
- 과소 적합은 훈련 자체를 너무 적게한 상태이므로 훈련 데이터에 대해서도 정확도가 낮다
- 머신 러닝에서 학습 또는 훈련 == 적합(fitting)이라고도 부름
- 모델이 주어진 데이터에 대해서 적합해져가는 과정
- 딥 러닝을 할 때는 과적합을 막을 수 있는 드롭 아웃(Dropout), 조기 종료(Early Stopping)과 같은 몇 가지 방법이 존재
---

[과적합 방지를 고려한 일반적인 딥 러닝 모델의 학습 과정]

1. 주어진 데이터를 훈련 데이터, 검증 데이터, 테스트 데이터로 나눈다. 가령, 6:2:2 비율로 나눌 수 있다.
2. 훈련 데이터로 모델을 학습한다. (에포크 +1)
3. 검증 데이터로 모델을 평가하여 검증 데이터에 대한 정확도와 오차(loss)를 계산한다.
4. 검증 데이터의 오차가 증가하였다면 과적합 징후이므로 학습 종료 후 Step 5로 이동, 아니라면 Step 2.로 재이동한다.
5. 모델의 학습이 종료되었으니 테스트 데이터로 모델을 평가한다.

## 03) 선형 회귀(Linear Regression)

### 1.선형 회귀(Linear Regression)

어떤 변수의 값에 따라서 특정 변수의 값이 영향을 받고 있다 <br>
값을 변하게하는 변수를 x, 변수 에 의해서 값이 종속적으로 변하는 변수 y <br>
- 독립변수 x, 종속변수 y
- 선형 회귀는 한 개 이상의 독립 변수 x와 y의 선형 관계를 모델링

#### 1) 단순 선형 회귀 분석(Simple Linear Regression Analysis)

$$
y = wx + b
$$

- 독립 변수 x 
- 곱해지는 값 w == 가중치(weight) == 직선의 기울기
- 별도로 더해지는 값 b == 편향(bias) == 절편
- x와 y의 값에 따라서 와 가 표현하는 직선은 무궁무진

#### 2) 다중 선형 회귀 분석(Multiple Linear Regression Analysis)

$$
y = w_1x_1 + w_2x_2 + ... + w_nx_n+ b
$$
- x가 여러개

### 3.비용 함수(Cost function) : 평균 제곱 오차(MSE)

머신 러닝은 w와 b를 찾기 위해서 실제값과 가설로부터 얻은 예측값의 오차를 계산하는 식을 세우고, 이 식의 값을 최소화하는 최적의 w와 b를 찾아낸다. <br>
실제값과 예측값에 대한 오차에 대한 식을 `목적 함수(Objective function) 또는 비용 함수(Cost function) 또는 손실 함수(Loss function) ` 라고 한다.
- 목적 함수 : 함수의 값을 최소화하거나, 최대화하거나 하는 목적을 가진 함수
- 비용 함수, 손실 함수 : 값을 최소화하려고 함, 예측값의 오차를 줄이는 일에 최적화 된 식 <br>

### 1.자동 미분

````python
import tensorflow as tf
````

텐서(tensor)란 일종의 행렬, 이는 스칼라가 될 수도 있고, 다차원의 벡터가 될 수 있다.
> 텐서플로(tensorflow) = 텐서(tensor)의 흐름(flow)를 설계


$$
2w^2 +5
$$

````python
# Variable은 변수
w = tf.Variable(2.)

def f(w):
  y = w**2
  z = 2*y + 5
  return z
````

````python
# 중간 연산 과정(함수, 연산)을 테이프(tape)에 차곡차곡 기록
with tf.GradientTape() as tape:
  z = f(w)

# w에 대해 미분한 값
gradients = tape.gradient(z, [w])
print(gradients)
````

### 2.자동 미분을 이용한 선형 회귀 구현

````python
# 학습될 가중치 변수를 선언
w = tf.Variable(4.0)
b = tf.Variable(1.0)
````

````python
# 가설을 함수로서 정의
@tf.function
def hypothesis(x):
  return w*x + b
````

````python
# 임의의 입력값을 넣었을 때
x_test = [3.5, 5, 5.5, 6]
print(hypothesis(x_test).numpy())
````

````python
# 평균 제곱 오차를 손실 함수로서 정의
@tf.function
def mse_loss(y_pred, y):
  # 두 개의 차이값을 제곱을 해서 평균을 취한다.
  return tf.reduce_mean(tf.square(y_pred - y))
````

````python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9] # 공부하는 시간
y = [11, 22, 33, 44, 53, 66, 77, 87, 95] # 각 공부하는 시간에 맵핑되는 성적
````

````python
# 옵티마이저는 경사 하강법을 사용하되, 학습률(learning rate)는 0.01을 사용
optimizer = tf.optimizers.SGD(0.01)
````

````python
# 약 300번에 걸쳐서 경사 하강법을 수행
for i in range(301):
  with tf.GradientTape() as tape:
    # 현재 파라미터에 기반한 입력 x에 대한 예측값을 y_pred
    y_pred = hypothesis(x)

    # 평균 제곱 오차를 계산
    cost = mse_loss(y_pred, y)

  # 손실 함수에 대한 파라미터의 미분값 계산
  gradients = tape.gradient(cost, [w, b])

  # 파라미터 업데이트
  optimizer.apply_gradients(zip(gradients, [w, b]))

  if i % 10 == 0:
    print("epoch : {:3} | w의 값 : {:5.4f} | b의 값 : {:5.4} | cost : {:5.6f}".format(i, w.numpy(), b.numpy(), cost))
````

w와 b값이 계속 업데이트 됨에 따라서 cost가 지속적으로 줄어든다.

````python
# 학습된 w와 b의 값에 대해서 임의 입력을 넣었을 경우의 예측값을 확인
x_test = [3.5, 5, 5.5, 6]
print(hypothesis(x_test).numpy())
````

### 2.시그모이드 함수(Sigmoid function)

시그모이드 함수는 종종 σ로 축약해서 표현
- 로지스틱 회귀를 풀기 위한 가설 세움

시그모이드 함수를 그래프로 시각화

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# w = 1, b= 0임을 가정한 그래프
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, 'g')
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
````

````python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x, y1, 'r', linestyle='--') # w의 값이 0.5일때 빨강
plt.plot(x, y2, 'g') # w의 값이 1일때 초록
plt.plot(x, y3, 'b', linestyle='--') # w의 값이 2일때 파랑
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
````

선형 회귀에서 직선을 표현할 때, 가중치 w는 직선의 기울기를 의미했지만 <br>
여기서는 그래프의 경사도를 결정한다. w의 값이 커지면 경사가 커지고 w의 값이 작아지면 경사가 작아진다.

````python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x, y1, 'r', linestyle='--') # x + 0.5 빨강
plt.plot(x, y2, 'g') # x + 1 초록
plt.plot(x, y3, 'b', linestyle='--') # x + 1.5 파랑
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
````

시그모이드 함수는 입력값이 커지면 1에 수렴하고, 입력값이 작아지면 0에 수렴한다.
